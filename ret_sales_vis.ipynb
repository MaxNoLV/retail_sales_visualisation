```python
# Identify and count entries with a letter in the invoice_number column

mask = data['invoice_number'].astype(str).str.match(r'^\d{7}[A-Za-z]$')
count_strange_instances = mask.sum()
print("Number of entries with 7 digits followed by a letter:", count_strange_instances)
```

```python
strange_instances = data[mask]
strange_instances
```

```python
strange_instances['invoice_number'].is_unique
```

```python
duplicate_mask = strange_instances['invoice_number'].duplicated(keep=False)
duplicates = strange_instances[duplicate_mask]
duplicates
```
